```python
##When we have only one feature (f1) it is called univariate analysis
##When we have only two features(f1,f2) it is called bivariate analysis
##univariate and bivariate analysis can be easily done by seaborn
##for multivariate we will be using various kinds of plots
##like distplot,joinplot,pairplot
##distplot helps us create a histogram 
##histogram shows density(count) of a feature based on x axis
##joinplot will help us do bivariate analysis
##from 3 onwards pairplot
##from all the features the dependant is the one for which we want to predict values



##Covariance is an indicator of the extent to which 2 random variables are dependent on each other.
##A higher number denotes higher dependency. 
##Correlation is a statistical measure that indicates how strongly two variables are related.
```

```python
import seaborn as sns
```

```python
##if our values are int or float then only we will be able to find correlation
##if our values are categorical features(string or objects) then well not be able to find correlation
##correlation value will be ranging from -1 to +1
```

```python
##if feature is categorical feature we have
##boxplot
##violinplot
##count plot
##bar plot
```

```python
ds=sns.load_dataset("tips")
```

```python
ds.head()
##tip is dependant feature all other are independent
```

```python
ds.corr(numeric_only=True)
ds.corr(numeric_only=True).dtypes
```

```python
sns.heatmap(ds.corr(numeric_only=True))
```

```python
sns.jointplot(x='tip',y='total_bill',data=ds,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=ds,kind='reg')
##draws a probability density function
##also draws a regression line (best fit line)
```

```python
sns.pairplot(ds)
```

```python
##visualise categorical data 
sns.pairplot(ds,hue='sex')
```

```python
ds['smoker'].value_counts()
```

```python
sns.distplot(ds['total_bill'])
```

```python
sns.distplot(ds['total_bill'],kde=True,bins=10) ##kde=kernel density estimation
```

```python
sns.countplot(x='sex',data=ds)
##in countplot we either have to give the x value or the y value
##other axis will always be giving the count
```

```python
sns.countplot(x='day',data=ds)
```

```python
sns.countplot(y='sex',data=ds)
```

```python
sns.barplot(y='sex',x='total_bill',data=ds)
```

```python
sns.boxplot(y="sex",x="total_bill",data=ds)
```

```python
sns.boxplot(y="sex",x="total_bill",data=ds,palette="rainbow")
```

```python
sns.boxplot(y="day",x="total_bill",data=ds)
```

```python
sns.boxplot(data=ds)
```

```python
sns.boxplot(data=ds,orient='v') ##by default it takes all numeric features
```

```python
sns.boxplot(y="day",x="total_bill",data=ds,hue="smoker") ##classify days vs total bill with repect to smoker yes or no
```

```python
sns.violinplot(x='day',y='total_bill',data=ds)
```
